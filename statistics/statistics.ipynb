```python
import pandas as pd
```

```python
df_stat = pd.read_csv('../data/df_cleaned_antibody.csv')
df_stat_copy = df_stat.copy()
df_stat_copy.head()
```

```python
columns = ['MIC (µg/mL)', 'MIC (µmol/mL)', 'log MIC', 'molar_mass']

for col in columns:
    print(f"Столбец: {col}")
    print(f"Медиана: {df_stat_copy[col].median()}")
    print(f"Мода: {df_stat_copy[col].mode()[0]}")
    print(f"Дисперсия: {df_stat_copy[col].var()}")
    print(f"Коэффициент вариации: {(df_stat_copy[col].std() / df_stat_copy[col].mean()) * 100}%")
    print(f"Ассиметрия: {df_stat_copy[col].skew()}")
    print(f"Эксцесс: {df_stat_copy[col].kurtosis()}")
    print(f"Амплитуда: {df_stat_copy[col].max() - df_stat_copy[col].min()}")
    print(f"Межквартильный размах: {df_stat_copy[col].quantile(0.75) - df_stat_copy[col].quantile(0.25)}")
    print("\n")
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
columns = ['MIC (µg/mL)', 'MIC (µmol/mL)', 'log MIC', 'molar_mass'] # Проверка распределения по тесту Шапиро

for col in columns:
    shapiro_test = scipy.stats.shapiro(df_stat_copy[col])
    if shapiro_test[1] < 0.05: 
        print(f"Данные распределены ненормально в столбце {col}")
    else:
        print(f"Данные распределены нормально в столбце {col}")
```

```python
# Создание боксплота

columns = ['MIC (µg/mL)', 'MIC (µmol/mL)', 'log MIC', 'molar_mass']

for col in columns:
    fig = px.box(df_stat_copy[col])
    fig.show()
```

```python
bacteria_list = df_stat_copy['bacteria'].unique()

for bacteria in bacteria_list:
    subset = df_stat_copy[df_stat_copy['bacteria'] == bacteria]
    fig = px.box(subset, x="bacteria", y="log MIC", title=f"Bacteria: {bacteria}")
    fig.show()
```

```python
# Создаем скрипичный плот

columns = ['MIC (µg/mL)', 'MIC (µmol/mL)', 'log MIC', 'molar_mass']

for col in columns:
    fig = px.violin(df_stat_copy[col])
    fig.show()
```

```python
# Матрица корреляции 

df_stat_heatmap = df_stat_copy[['MIC (µg/mL)', 'MIC (µmol/mL)', 'log MIC', 'molar_mass']]
correlation=df_stat_heatmap.corr()
df_stat_heatmap.corr()
```

```python
# Heatmap

fig = px.imshow(
    correlation,
    color_continuous_scale='twilight',  # Можно изменить цветовую схему, например на 'Blues', 'Inferno', 'Viridis'
    labels=dict(color="Correlation"),  # Добавить подпись к шкале
    aspect="auto",  # Автоматически подстроить соотношение сторон
    zmin=-1,  # Минимальное значение шкалы корреляции
    zmax=1    # Максимальное значение шкалы корреляции
)

# Добавить аннотации для значений корреляции
fig.update_traces(text=correlation.values.round(2), texttemplate="%{text}", textfont_size=12)

# Настроить внешний вид
fig.update_layout(
    title="Correlation Heatmap",  # Заголовок графика
    title_x=0.5,  # Центрировать заголовок
    width=800,  # Ширина графика
    height=600,  # Высота графика
    coloraxis_colorbar=dict(
        title="Correlation",  # Подпись к шкале цветов
        tickvals=[-1, 0, 1],  # Отображаемые значения на шкале
        ticktext=["-1", "0", "1"]
    )
)
```
