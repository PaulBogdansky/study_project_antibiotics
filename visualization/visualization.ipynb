```python
ds_sort = ds.sort_values(by='Molecular weight').head(100) 
```

```python
import pandas as pd
import plotly.graph_objs as go


#  3D scatter plot
scatter_3d = go.Scatter3d(
    x=ds['Molecular weight'],  #  X
    y=ds['MIC (mol mL1)'],   #  Y
    z=ds['log MIC'],           #  Z
    mode='markers',
    marker=dict(
        size=5,
        color=ds['Molecular weight'],  #      
        colorscale='Viridis',  #  
        opacity=0.8
    )
)

#   
layout = go.Layout(
    title='3D Scatter Plot: Molecular weight vs MIC (mol mL1) vs log MIC',
    scene=dict(
        xaxis=dict(title='Molecular weight'),
        yaxis=dict(title='MIC (mol mL1)'),
        zaxis=dict(title='log MIC')
    )
)

#    
fig = go.Figure(data=[scatter_3d], layout=layout)
fig.show()
```

```python
MIC_mol_mL_1 = "MIC (mol mL1)"

data = ds[MIC_mol_mL_1]
hist = go.Histogram(x=data, nbinsx=40, name='Histogram', marker=dict(color='skyblue', line = dict(color='black', width = 1.5)))
layout = go.Layout(bargap=0.2, title='   MIC (mol mL1)')
fig = go.Figure(data=[hist], layout=layout)

fig.show()
#  
```

```python
filtered_data.loc[:, 'MIC (mol mL1)'] = pd.to_numeric(filtered_data['MIC (mol mL1)'], errors='coerce')
filtered_data = filtered_data.dropna(subset=['MIC (mol mL1)'])

#   ,   NaN   -   !!!!!
```

```python
mean = filtered_data['MIC (mol mL1)'].mean()
std_dev = filtered_data['MIC (mol mL1)'].std()
x = np.linspace(filtered_data['MIC (mol mL1)'].min(), filtered_data['MIC (mol mL1)'].max(), 100)
y = norm.pdf(x, mean, std_dev) #           
plt.plot(x, y, label=" ", color="red") #  -   


#        

# :     ,     
```
