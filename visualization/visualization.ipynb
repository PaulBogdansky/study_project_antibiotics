```python
import pandas as pd
import plotly.graph_objs as go


# Построение 3D scatter plot
scatter_3d = go.Scatter3d(
    x=ds['Molecular weight'],  # Ось X
    y=ds['MIC (µmol mL−1)'],   # Ось Y
    z=ds['log MIC'],           # Ось Z
    mode='markers',
    marker=dict(
        size=5,
        color=ds['Molecular weight'],  # Цвет точек основан на весе молекул
        colorscale='Viridis',  # Цветовая схема
        opacity=0.8
    )
)

# Определение макета графика
layout = go.Layout(
    title='3D Scatter Plot: Molecular weight vs MIC (µmol mL−1) vs log MIC',
    scene=dict(
        xaxis=dict(title='Molecular weight'),
        yaxis=dict(title='MIC (µmol mL−1)'),
        zaxis=dict(title='log MIC')
    )
)

# Создание и отображение графика
fig = go.Figure(data=[scatter_3d], layout=layout)
fig.show()
```

```python
sns.histplot(filtered_data, bins=20, kde=False, stat="density", label="Гистограмма")

# Строим гистограмму так, чтобы по ординат у нас была плотность. kde=False - вырубаем встроенную кривую плотности, ибо будем добавлять собственный колокол
```

```python
filtered_data.loc[:, 'MIC (µmol mL−1)'] = pd.to_numeric(filtered_data['MIC (µmol mL−1)'], errors='coerce')
filtered_data = filtered_data.dropna(subset=['MIC (µmol mL−1)'])

# избавляемся от строк, переводим в NaN и дропаем - иначе не работает!!!!!
```

```python
mean = filtered_data['MIC (µmol mL−1)'].mean()
std_dev = filtered_data['MIC (µmol mL−1)'].std()
x = np.linspace(filtered_data['MIC (µmol mL−1)'].min(), filtered_data['MIC (µmol mL−1)'].max(), 100)
y = norm.pdf(x, mean, std_dev) # Считаем значение плотности норм распределения для каждой точки на оси Х
plt.plot(x, y, label="Нормальное распределение", color="red") # Рисуем колокол-лайк кривую норм распределения


# Вычисляем среднее и стандартное отклонение для нормального распредления

# х: Создаем диапазон значений для абцисс, чтобы была гладкая линия 
```
