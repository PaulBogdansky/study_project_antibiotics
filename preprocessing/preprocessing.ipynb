```python
from preprocessing import (
    find_project_root, load_dataset_from_project, clean_dataset, contains_cyrillic, 
    detect_cyrillic_in_dataframe, replace_cyrillic, calculate_molar_mass, 
    validate_and_correct_MIC, standardize_features

)
```

```python
project_name = 'study_project_antibiotics'
file_path = 'data/antibiotics_row.csv'

df = load_dataset_from_project(project_name, file_path)
```

```python
columns_to_drop = ['doi', 'id', 'type', 'group', 'group_name', 'Unnamed: 9', 'E. coli', 'Unnamed: 15', 'MIC (µmol mL−1)']

df_cleaned = clean_dataset(df, columns_to_drop)

detect_cyrillic_in_dataframe(df_cleaned)
```

```python
df_cleaned['molar_mass'] = df_cleaned['SMILES'].apply(calculate_molar_mass)

df_cleaned_validate = validate_and_correct_MIC(df_cleaned)

df_cleaned = validate_and_correct_MIC(df_cleaned)
```

```python
numerical_features = ['MIC (µg/mL)', 'MIC (µmol/mL)', 'log MIC', 'molar_mass']

df_cleaned_std =  standardize_features(df_cleaned, numerical_features)
```

```python
df_cleaned_validate.to_csv('../data/df_cleaned_antibody.csv', index=False)

df_cleaned_std.to_csv('../data/df_cleaned_antibody_std.csv', index=False)
```
